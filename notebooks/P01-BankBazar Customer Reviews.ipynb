```python
import requests # for sending requests 
from bs4 import BeautifulSoup # scraping data
import pandas as pd # datasets building
```

## Send Request

```python
# send request to the web server 
url = "https://www.bankbazaar.com/reviews.html"
page = requests.get(url) 

# parse data 
soup = BeautifulSoup(page.text, "html.parser")
```

## Reading Data 

```python
# read review text 
review_text = []
review_text_elem = soup.find_all(class_='text_here review-desc-more')
for item in review_text_elem: 
    review_text.append(item.text.replace("\n", "").replace("\t", "").strip())
```

```python
review_text 
```

```python
# read username 
user_name = [] 
user_name_elem = soup.find_all(class_='js-author-name')
for item in user_name_elem: 
    user_name.append(item.text)
```

```python
user_name 
```

```python
bank_name = [] 
bank_name_elem = soup.find_all('div', class_='review-bank-title')
for item in bank_name_elem: 
    bank_name.append(item.find('img').get('alt')) 
```

```python
bank_name 
```

## Creating Dataset

```python
# method-1 
df = pd.DataFrame(list(zip(review_text, user_name, bank_name)), columns=["Reviews", "User", "Bank"])
df 
```

```python
# method-2 
df = pd.DataFrame(
    {'Reviews': review_text, 
    'User': user_name, 
    'Bank': bank_name}
)

df 
```

## Exporting Data 

```python
# export data as csv 
df.to_csv("../data/bank_bazar_reviews.csv", index=False)
```
